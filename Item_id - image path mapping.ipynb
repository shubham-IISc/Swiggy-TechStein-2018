```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_files
import pickle
```

```python
#reading given items_metadata.csv provided to us
df_item=pd.read_csv('items_metadata.csv')
df_item.head(10)
```

```python
#loading images provided to us corresponding to differnt dish_family
bunch=load_files('static/Img',shuffle=False)
```

```python
file_names=list(bunch.filenames)
```

```python
df_item=df_item.sort_values(by='dish_family')
```

```python
item=list(df_item['dish_family'])
```

```python
i=0
j=0
final_item=[]
final_file=[]
l_item=len(item)
l_file=len(file_names)
```

```python
print(l_item)
print(l_file)
print(file_names[:10])
```

```python
#assigning each Item_id an image corresponding to its given dish_family
while(i<l_item):
    if(item[i][:3]==file_names[j][11:14]):
        final_item.append(item[i])
        final_file.append(file_names[j])
        i=i+1
        j=j+1
        if(j==l_file):
            j=0
    else:
        j=j+1
        if(j==l_file):
            j=0
            
```

```python
#finally creating a dataframe with 'item' (Item_id) as one column and 'image'(corresponding image path) as other
image_lookup=pd.DataFrame({})
image_lookup['item']=final_item
image_lookup['image']=final_file
```
